```python
                                                            TABLE PLOT
```

```python

a = load_iris()
a_df = pd.DataFrame(data=np.c_[a['data'],
                                  a['target']], 
                       columns=a['feature_names'] + ['target'])
grouped_dataframe = a_df.groupby('target').mean().round(1)
grouped_dataframe['species_name'] = ['setosa', 'versicolor', 'virginica']
ax = plt.subplot(211)
plt.title("Iris Dataset Average by Plant Type")
plt.ylabel('Centimeters (cm)')
ticks = [4, 8, 12, 16]
b = [x - 1 for x in ticks]
c = [x + 1 for x in ticks]
plt.xticks([])
plt.bar(b, grouped_dataframe.loc[0].values.tolist()[
        :-1], width=1, label='setosa')
plt.bar(ticks, grouped_dataframe.loc[1].values.tolist()[
        :-1], width=1, label='versicolor')
plt.bar(c, grouped_dataframe.loc[2].values.tolist()[
        :-1], width=1, label='virginica')
plt.legend()
plt.figure(figsize=(12, 8))
table(ax, grouped_dataframe.drop(['species_name'], axis=1), loc='bottom')
```

```python
                                                            PIE CHART
```

```python
import pandas as pd
import matplotlib.pyplot as plt
a = pd.read_csv("Iris.csv")
ax=plt.subplots(1,1,figsize=(10,8))
a['Species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.title("Iris Species ")
plt.show()
```
